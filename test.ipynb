```python
import networkx as nx
g = nx.Graph()
```

```python
# Adding multiple nodes
g.add_nodes_from(['Hillary', 'Michelle', 'Obama', 'Trump'])

g.nodes()
```

```python
g.add_edges_from([
    ('Obama', 'Michelle'),
    ('Hillary', 'Clinton'),
    ('Obama', 'Trump'),
    ('Obama', 'Clinton'),
])

g.edges()
```

```python
nx.info(g)
```

```python
nx.draw(g, with_labels=True)
```

```python
nx.draw_networkx(g, directed=True)
```

```python
nx.eigenvector_centrality(g)
```

```python
nx.betweenness_centrality(g)
```
